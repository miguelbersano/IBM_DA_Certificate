#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `RemoteWork`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
#let's first see column data types and number of missing values
missing_values = df.isnull()
df_columns = df.columns.values.tolist()
for column in df_columns:
    print(column)
    print('type: ', df[column].dtypes, '  count: ', df[column].count(), '  missing values:', missing_values[column].sum(), '  % of missing values: ', 
          f'{(missing_values[column].sum() / len(df.index)) * 100: .2f}%', '\n')
```

```python
#Employment does not have missing values
#RemoteWork has 16.25% of missing values, let's replace them by the most frequent value
RemoteWork_mode = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].replace(np.nan, RemoteWork_mode)
#JobSat has a high percentage of missing values (55.49%); I don't think that I should adopt a strategy to deal with them right now. I will deal with them when an analysis involving JobSat is required; I 
#will then devise the most appropriate strategy, depending on the type of the required analysis 
#let's impute missing values (most frequent) to DevType, as we will be using this column on later analyses; it contains around 9% of missing values
DevType_mode = df['DevType'].mode()[0]
df['DevType'] = df['DevType'].replace(np.nan, DevType_mode)
```

#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
#let's see the sample size if we drop all rows with missing values for in one of the columns
df_ycpjs = df[['YearsCodePro', 'JobSat']].dropna()
df_ycpjs.shape
```

+++ We have 28356 rows in this sample, it is a good size to do this analysis, so let's do it with this strategy for dealing with missing values; let's create a column with experience range and see its
relationship with JobSatisfaction


```python
#lets define a function to assign the experience levels
def exp_range(years_code_pro):
    exp_ranges = {5: '0-5', 10: '6-10', 20: '11-20', 30: '21-30', 100: 'More than 30'}

    if pd.isna(years_code_pro):
        return np.nan
    if years_code_pro == 'Less than 1 year':
        return exp_ranges[5]
    if years_code_pro == 'More than 50 years':
        return exp_ranges[100]

    try:
        years = float(years_code_pro)
    except (ValueError, TypeError):
        return np.nan

    for limit, exp in exp_ranges.items():
        if years <= limit:
            return exp
```

```python
#let's create the column
df_ycpjs['ExperienceRange'] = df_ycpjs['YearsCodePro']. apply(exp_range)
df_ycpjs.head()
```

```python
df_erjs = df_ycpjs.groupby('ExperienceRange')['JobSat'].mean().reset_index()
df_erjs.rename(columns={'JobSat': 'MeanJobSat'}, inplace=True)
```

```python
# Create the bar chart using seaborn - average JobSat
plt.figure(figsize=(10,6))
sns.barplot(x='ExperienceRange', y='MeanJobSat', data=df_erjs)
plt.xlabel('Experience Range (years)')
plt.ylabel('Average Job Satisfaction')
plt.title('Average Job Satisfaction per Experience Range')
plt.show()
```

```python
#let's plot using median as estimator, without using .groupby
plt.figure(figsize=(10,6))
sns.barplot(x='ExperienceRange', y='JobSat', data=df_ycpjs, estimator=np.median)
plt.xlabel('Experience Range (years)')
plt.ylabel('Median Job Satisfaction')
plt.title('Median Job Satisfaction per Experience Range')
plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
#count plot for RemoteWork
plt.figure(figsize=(10,6))
sns.countplot(x='RemoteWork', data=df, color='lightblue')
plt.xlabel('Remote Work')
plt.ylabel('Count')
plt.title('Remote Work count distribution')
plt.show()
```

```python
#let's see first the remote work preferences for employment type
#let's determine the original unique options for Employment in the survey
emp_options = df['Employment'].dropna().str.split(';').explode().unique()
#let's tabulate RemoteWork preferences by employment type; for this, we will first create a boolean column for each original employment option in the survey
for et in emp_options:
    df[et] = df['Employment'].str.contains(et, na=False)
#now let's tabulate
summary = df.groupby('RemoteWork')[
    ['Employed, full-time',
     'Student, full-time',
     'Not employed, but looking for work',
     'Independent contractor, freelancer, or self-employed',
     'Not employed, and not looking for work',
     'Student, part-time',
     'Employed, part-time',
     'I prefer not to say',
     'Retired']
].sum()
summary
```

```python
#we can plot a heatmap showing this tabulation
plt.figure(figsize=(10,5))
sns.heatmap(summary, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Remote Work Preferences by Employment Type')
plt.ylabel('Remote Work')
plt.xlabel('Employment Type')
plt.xticks(rotation=45)
plt.show()
```

```python
#let's now look at the relationship of the three variables (RemoteWork, Employment and DevType (job role)
summary_1 = df.groupby(['RemoteWork', 'DevType'])[
    ['Employed, full-time',
     'Student, full-time',
     'Not employed, but looking for work',
     'Independent contractor, freelancer, or self-employed',
     'Not employed, and not looking for work',
     'Student, part-time',
     'Employed, part-time',
     'I prefer not to say',
     'Retired']
].sum()
summary_1
```

```python
#we can plot a heatmap showing this tabulation
plt.figure(figsize=(10,20))
sns.heatmap(summary_1, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Remote Work Preferences by DevType and Employment Type')
plt.ylabel('Remote Work by DevType')
plt.xlabel('Employment Type')
plt.xticks(rotation=45)
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
#To do this analysis, let's drop rows with missing values for columns Country and LanguageHaveWorkedWith
df_clhww = df[['Country', 'LanguageHaveWorkedWith']].dropna()
```

```python
#let's see the original individual language options in the survey
language_options = df_clhww['LanguageHaveWorkedWith'].str.split(';').explode().unique()
len(language_options)
```

```python
#let's expand the dataframe by creating one row for each language when we have multiple options in LanguageHaveWorkedWith
df_clhww['Language'] = df_clhww['LanguageHaveWorkedWith'].str.split(';') # create a new column with a list of all language options
df_clhww.head()
```

```python
df_clhww = df_clhww.explode('Language') # replicate rows with more than one language in this new column
df_clhww.head(10)
```

```python
df_clhww['Country'].nunique()
```

```python
#as we have 185 countries and 49 languages, we cannot plot a single graph showing this relationship; we will have to analyze by parts...
#Filter to Top Languages and Selected Countries
top_languages = df_clhww['Language'].value_counts().head(10).index

selected_countries = ['United States of America', 'India', 'Germany']

filtered_df_clhww = df_clhww[
    (df_clhww['Language'].isin(top_languages)) &
    (df_clhww['Country'].isin(selected_countries))
]

filtered_df_clhww
```

```python
#We can plot a FacetGrid plot for these filtered countries and languages
g = sns.catplot(
    data=filtered_df_clhww,
    x='Language',
    kind='count',
    col='Country',
    col_wrap=3,
    order=top_languages,
    height=8,
    aspect=1.2
)

g.set_xticklabels(rotation=45)

# Set axis labels font size
g.set_xlabels('Programming Language', fontsize=14)
g.set_ylabels('Number of Developers', fontsize=14)

# Set tick labels font size
for ax in g.axes.flat:
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)

# Set suptitle font size
g.fig.suptitle('Top Programming Languages by Selected Countries', fontsize=18, y=1.02)

plt.show()
```

```python
#we can do a cross-tabulation of selected countries and top languages
top_languages = df_clhww['Language'].value_counts().head(10).index
selected_countries = ['United States of America', 'India', 'Germany', 'Brazil']
filtered_df_clhww = df_clhww[
    (df_clhww['Language'].isin(top_languages)) &
    (df_clhww['Country'].isin(selected_countries))]
lang_pref_by_country = pd.crosstab(filtered_df_clhww['Language'], filtered_df_clhww['Country'])

lang_pref_by_country
```

```python
#we can plot a heatmap showing this tabulation
plt.figure(figsize=(10,20))
sns.heatmap(lang_pref_by_country, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Top Language preferences by selected Country')
plt.ylabel('Language')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
#let's see the sample size if we drop all rows with missing values in one of the columns
df_ycpjsp_1 = df[['YearsCodePro', 'JobSatPoints_1']].dropna()
df_ycpjsp_1.shape
```

+++We have 28496 rows in the sample, it is a good size to do this analysis

```python
# Map the special string values to numeric equivalents
mapping = {
    'Less than 1 year': 0,
    'More than 50 years': 51
}

# Convert YearsCodePro to numeric values, replacing special strings
df_ycpjsp_1['YearsCodeProNum'] = df_ycpjsp_1['YearsCodePro'].replace(mapping).astype(float)

# Sort by the numeric version
#we do this because in matplotlib, when the x-values are categorical, they will appear on the x-axis on the order they appear for the first time in a row
df_ycpjsp_1 = df_ycpjsp_1.sort_values('YearsCodeProNum')

# Plot using the sorted numeric x-axis but keep original labels
plt.figure(figsize=(20, 10))
plt.scatter(df_ycpjsp_1['YearsCodePro'], df_ycpjsp_1['JobSatPoints_1'])
plt.xticks(rotation=60, ha='right')
plt.title('Job Satisfaction Points (1) by Years of Professional Coding Experience')
plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
#let's first create a new dataframe for this analysis, dropping all rows which have missing values for these columns and exploding rows with multiple options for Employment
df_ele = df[['EdLevel', 'Employment']].dropna()
#let's expand the dataframe by creating one row for each  employment type when we have multiple options in Employment
df_ele['EmploymentType'] = df_ele['Employment'].str.split(';') # create a new column with a list of all employment options
df_ele = df_ele.explode('EmploymentType') # replicate rows with more than one employment type in this new column
df_ele.head(10)
```

```python
#let's cross-tabulate employment type by education level
emptype_by_edlevel = pd.crosstab(df_ele['EmploymentType'], df_ele['EdLevel'])
emptype_by_edlevel
```

```python
from textwrap import fill
```

```python
#plotting a heatmap with this cross-tabulation
plt.figure(figsize=(20,6))
ax = sns.heatmap(
    emptype_by_edlevel,
    annot=True,
    fmt='d',
    cmap='YlGnBu',
    linewidths=0.3,
    cbar_kws={'fraction': 0.046, 'pad': 0.04}
)

# Wrap long labels to fit nicely
wrapped_labels = [fill(l, 18) for l in emptype_by_edlevel.columns]

n_cols = emptype_by_edlevel.shape[1]
ax.set_xticks(np.arange(n_cols) + 0.5)
ax.set_xticklabels(
    wrapped_labels,
    rotation=45,
    ha='right',
    rotation_mode='anchor',
    fontsize=12   # ⬅ increase x-tick font size
)

ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)  # ⬅ increase y-tick font size

ax.set_xlim(0, n_cols)

plt.subplots_adjust(bottom=0.32, left=0.15, right=0.95, top=0.92)

# Titles and axis labels with larger font
ax.set_title('Employment type by Education Level', pad=16, fontsize=14, fontweight='bold')
ax.set_ylabel('Employment Type', fontsize=13)
ax.set_xlabel('Educational Level', labelpad=10, fontsize=13)

plt.show()
```
