Print the mean age of the survey participants.


'Age' is not a numnerical column, so we cannot calculate its mean. But we can plot a pie plot showing the percentages of responses for each age bracket. Will do this below.

```python
df['Age'].unique()
```

```python
df_ages = df.groupby(['Age'])['ResponseId'].count().reset_index()
df_ages
```

```python
no_of_rows = df_ages['ResponseId'].sum()
df_ages['Percentage'] = (df_ages['ResponseId'] / no_of_rows * 100).round(2)
df_ages
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

labels = list(df_ages['Age'])
sizes = list(df_ages['Percentage'])

# Draw pie without labels (only wedges)
wedges, _ = ax.pie(
    sizes,
    labels=None,
    startangle=90,
    wedgeprops=dict(edgecolor='white')
)

# Legend
ax.legend(wedges, labels, title="Age", loc='center left', bbox_to_anchor=(1, 0.6))

# Collect label positions
label_positions = []

for i, wedge in enumerate(wedges):
    angle = (wedge.theta2 + wedge.theta1) / 2.0
    x = np.cos(np.deg2rad(angle))
    y = np.sin(np.deg2rad(angle))
    ha = 'left' if x >= 0 else 'right'

    # Proposed position for text
    pos = np.array([x * 1.2, y * 1.2])

    label_positions.append([pos, sizes[i], ha])

# --- Simple overlap adjustment ---
# Sort by y-coordinate to avoid collisions
label_positions.sort(key=lambda t: t[0][1])

min_gap = 0.08  # minimum vertical gap between labels

for j in range(1, len(label_positions)):
    y_prev = label_positions[j-1][0][1]
    y_curr = label_positions[j][0][1]
    if y_curr - y_prev < min_gap:
        # shift current label upward
        label_positions[j][0][1] = y_prev + min_gap

# Draw annotations after adjustment
for pos, val, ha in label_positions:
    x, y = pos
    ax.annotate(
        f"{val:.1f}%",
        xy=(x*0.7, y*0.7),   # arrow anchor near wedge
        xytext=(x, y),       # adjusted label position
        horizontalalignment=ha,
        verticalalignment='center',
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.2", fc="white", alpha=0.8),
        arrowprops=dict(arrowstyle='-', color='gray', lw=0.8)
    )

ax.set_title('Percentages for each age bracket', pad=30)
plt.tight_layout()
plt.show()
```

```python
#We can calculate an estimation of the average age, by assuming an average age for each age bracket, and calculating the weighted average of them ('Prefer not to say' not included)
#let's assume 13.5 years for the 'Under 18 years old' bracket and 72.5 years for the '65 years or older' bracket
#For the other brackets let's take the the average of the two bracket limits
#So:
df_ages_1 = df_ages.drop(6)
no_of_rows = df_ages_1['ResponseId'].sum()
df_ages_1['Percentage'] = (df_ages_1['ResponseId'] / no_of_rows * 100).round(2)
bracket_avgs = [21, 29.5, 39.5, 49.5, 59.5, 72.5, 13.5]
df_ages_1['Bracket_avg_age'] = bracket_avgs
df_ages_1['Weighted_Bracket_avg_age'] = df_ages_1['Bracket_avg_age'] * df_ages_1['Percentage'] / 100
df_ages_1
print('The estimated average age is: ', df_ages_1['Weighted_Bracket_avg_age'].sum().round(1), ' years')
```
